# Regresión Lineal Simple
 ## Utilizando Least mean square (mínimos cuadrados)
 ## Loss function: MSE

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# Genera dataset para regresión
from sklearn.datasets import make_regression
```

```python
# Crea el dataset 
X_R1, y_R1 = make_regression(n_samples = 100, n_features=1,
                            n_informative=1, bias = 150.0,
                            noise = 30, random_state=0)
# Divsión de training-test set
X_train, X_test, y_train, y_test = train_test_split(X_R1, y_R1,
                                                   random_state = 0)
# Calculará los coeficientes dado un conjunto de datos Entrenamiento
linreg = LinearRegression().fit(X_train, y_train)


print('linear model coeff (w): {}'
     .format(linreg.coef_))
print('linear model intercept (b): {:.3f}'
     .format(linreg.intercept_))
print('R-squared score (training): {:.3f}'
     .format(linreg.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linreg.score(X_test, y_test)))
```

### Mean squared Error

```python
def mse(y_hat, y): 
    return ((y_hat - y)**2).mean()
```

**Tenemos el vector a=(-1,1) después calcula y_hat(estimada) y calcula el error**

```python
# Generamos el vector
a = np.array([-1.,1])
# CAlculamos el MSE
y_hat = x.dot(a)
mse(y_hat, y)
```

```python
# Este es el arreglo de MSE
(y - y_hat)**2 / n
#print(np.shape(mse))
```

## Descenso de Gradiente 
#### Se encarga de encontrar los valores que minimicen la función de error.
Inicia con un set inicial de parámetros e iterartivamente se dirige hacia otro set de parámetros para minimizar la función objetivo

```python
def derivatives2d():
    global y_hat, y
    error = y - y_hat
    derivatives = -(1.0 / n) * error.dot(x)
    return derivatives, mse(y, y_hat)
```

```python
derivatives2d()
```

#### Updating the model coefficientes

```python
def update():
    global a, lr, t, y_hat
    y_hat = x.dot(a)
    grad, loss = derivatives2d()
    a = a - (lr * grad)
    if t % 10 == 0: 
        print(loss, grad)
    grad = np.zeros(2)
    return loss
```

```python
# learning rate .01 hasta 100, columna izquierda es el MSE y el array son los pesos que corresponden a ese MSE
lr = 1e-1
for t in range(100): 
    loss = update()
```

# Regresión Ridge

```python
from sklearn.linear_model import Ridge

X_train, X_test, y_train, y_test = train_test_split(X_R1, y_R1,
                                                   random_state = 0)
linridge = Ridge(alpha=20.0).fit(X_train, y_train)


print('ridge regression linear model intercept: {}'
     .format(linridge.intercept_))
print('ridge regression linear model coeff:\n{}'
     .format(linridge.coef_))
print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
```

```python
print(y_R1.shape,X_R1.shape, linridge.coef_.shape)
```

```python
plt.figure(figsize=(5,4))
plt.scatter(X_R1, y_R1, marker= '^', s=50, alpha=0.9)
plt.plot(X_R1, linridge.coef_ * X_R1 + linridge.intercept_, 'r-')
plt.title('Regresión Lineal con Penalización Ridge por cuadrados mínimos')
plt.xlabel('Valor de característica (x)')
plt.ylabel('Salida (y)')
plt.show()
```

# Regresión Lasso

```python
from sklearn.linear_model import Lasso

# Crea el dataset sintético
X_R2, y_R2 = make_regression(n_samples = 150, n_features=1,
                            n_informative=1, bias = 100.0,
                            noise = 30, random_state=0)

X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_R1, y_R1,
                                                   random_state = 0)

linlasso = Lasso(alpha=2.0, max_iter = 10000).fit(X_train_2, y_train_2)

print('Crime dataset')
print('lasso regression linear model intercept: {}'
     .format(linlasso.intercept_))
print('lasso regression linear model coeff:\n{}'
     .format(linlasso.coef_))
print('Non-zero features: {}'
     .format(np.sum(linlasso.coef_ != 0)))
print('R-squared score (training): {:.3f}'
     .format(linlasso.score(X_train_2, y_train_2)))
print('R-squared score (test): {:.3f}\n'
     .format(linlasso.score(X_test_2, y_test_2)))
```

```python
plt.figure(figsize=(5,4))
plt.scatter(X_R1, y_R1, marker= '^', s=50, alpha=0.9)
plt.plot(X_R1, linlasso.coef_ * X_R1 + linlasso.intercept_, 'r-')
plt.title('Regresión Lineal con Penalización Ridge')
plt.xlabel('Valor de característica (x)')
plt.ylabel('Salida (y)')
plt.show()
```
